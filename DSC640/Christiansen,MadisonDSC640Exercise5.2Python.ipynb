```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objs as go

df = pd.read_csv('/Users/madisonchristiansen/Desktop/birth-rate.csv')   
df.tail()
```

```python
# Histogram
plt.figure(figsize=(20, 10))
plt.hist(df, bins=10, edgecolor='pink')
plt.xticks(rotation=45)
plt.title('Histogram')
plt.xlabel('Country')
plt.ylabel('1960')
plt.show()
```

```python
# Box Plot
countryrow = df[df['Country'] == 'Samoa']

value = countryrow.iloc[:, 1:].values.flatten()

plt.figure(figsize=(8, 6))
plt.boxplot(value)
plt.title('Box Plot')
plt.xlabel('Value')
plt.show()
```

```python
# Bullet Chart
year = countryrow.columns[1:]

target = 20.0

d1 = plo.Bar(x=year, y=value, name='Actual', marker=dict())
d2 = plo.Scatter(x=year, y=[target] * len(year), mode='lines', name='Target', line=dict(color='black', dash='dash'))

layout = plo.Layout(
    title='Bullet Chart',
    xaxis=dict(title='Year'),
    yaxis=dict(title='Value'))

fig = plo.Figure(data=[d1, d2], layout=layout)
fig.show()
```

```python
# Line Chart
plt.figure(figsize=(15, 6))
plt.plot(year, value)
plt.xticks(rotation=45)
plt.grid(True)
plt.title('Line Chart')
plt.xlabel('Year')
plt.ylabel('Value')
plt.show()
```
