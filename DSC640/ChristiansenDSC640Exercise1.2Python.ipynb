### Exercise 1.2 - Python
#### 1 bar chart, 1 stacked bar chart, 1 pie chart, 1 donut, and 1 line chart with Python


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_excel('/Users/madisonchristiansen/Desktop/hotdog-contest-winners.xlsm')   
df.tail()
```

```python
## Bar Chart
Winner = df['Winner']
Dogs = df['Dogs eaten']
 
fig = plt.figure(figsize =(30, 10))
plt.bar(Winner, Dogs)
 
## Show Plot
plt.ylabel('Dogs Eaten')
plt.title('Dogs Eaten by Winner')
plt.show()
```

```python
## Stacked Bar Chart 
fig, ax = plt.subplots()
ax.bar(df.index, df['Dogs eaten'], label='Dogs eaten') # top bar
ax.bar(df.index, df['New record'], bottom=df['Dogs eaten'], # bottom bar
       label='New record')
## Show Plot
ax.set_title('Dogs Eaten and New Record')
ax.legend()
```

```python
## Pie Chart
df["Country"].value_counts().plot.pie(colors = ['pink', 'red', 'purple', 'lightblue'], autopct='%1.0f%%')
```

```python
## Donut Chart
avg_dogs_eaten = df.groupby('Country')['Dogs eaten'].mean() # average dogs by country

fig, ax = plt.subplots()

wedges, text, autotexts = ax.pie(avg_dogs_eaten, labels=avg_dogs_eaten.index, autopct='%1.1f%%')
ax.add_artist(plt.Circle((0, 0), 0.4, fc='white'))
ax.set_aspect('equal')
## Show Plot
ax.set_title('Average Number of Hot Dogs Eaten by Country')
plt.show()
```

```python
## Line Chart
fig, ax = plt.subplots()

ax.plot(df['Year'], df['Dogs eaten']) # dogs eaten by year
ax.set_xticks(df['Year'])
ax.set_xticklabels(df['Year'], rotation=45)
## Show Plot
ax.set_title('Number of Hot Dogs Eaten Over the Years')
ax.set_xlabel('Year')
ax.set_ylabel('Dogs Eaten')
plt.tight_layout()
plt.show()
```
