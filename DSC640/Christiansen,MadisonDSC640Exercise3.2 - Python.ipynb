### 1 scatterplot, 1 bubble chart and 1 density plot chart 


```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('/Users/madisonchristiansen/Desktop/crimerates-by-state-2005.csv')
df.tail()
```

```python
# Scatterplot 

plt.figure(figsize=(30, 15))
plt.scatter(df.state, df.robbery)
plt.show()
```

```python
# Bubble Chart - Burglaryvs Murder by Robbery
plt.figure(figsize=(10, 6))

plt.scatter(df['murder'], df['burglary'], s=df['robbery']*10, alpha=0.7, cmap='viridis') ## scatter

colorbar = plt.colorbar()
colorbar.set_label('Robbery')

plt.xlabel('Murder')
plt.ylabel('Burglary')
plt.title('Bubble Chart')
plt.show()
```

```python
import seaborn as sns

# Density Plot Chart 
plt.figure(figsize=(10, 6))

sns.kdeplot(data=df[['murder', 'burglary']], fill=True) ## density plot

plt.title('Density Plot')
plt.show()
```
