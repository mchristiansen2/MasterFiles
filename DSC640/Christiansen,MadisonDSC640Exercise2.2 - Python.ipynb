### 1 tree map, 1 area chart, 1 stacked area chart, and 1 step chart using Python

```python
import pandas as pd
import matplotlib.pyplot as plt
import squarify


df = pd.read_csv('/Users/madisonchristiansen/Desktop/unemployement-rate-1948-2010.csv')   
df.tail()
```

```python
# Treemap 
## Using squarify

labels = ["{year}-{period}" for year, period in zip(df['Year'], df['Period'])]
sizes = df['Value'] ## based on value

plt.figure(figsize=(25, 20))
squarify.plot(sizes=sizes, label=labels)

plt.title('Tree Map - Based on Value Column')
plt.show()
```

```python
# Area Chart - on vlaue column 

plt.figure(figsize=(15, 10))
plt.fill_between(df.index, df['Value'], alpha=0.5)
plt.ylabel('Value')
plt.title('Area Chart')
plt.show()
```

```python
# Stacked Area Chart
df_pivot = df.pivot(index='Year', columns='Period', values='Value')
plt.stackplot(df_pivot.index, df_pivot.values.T, labels=df_pivot.columns)

plt.xlabel('Year')
plt.ylabel('Value')
plt.title('Stacked Area Chart')
plt.legend(loc='upper left')
plt.show()
```

```python
# Step chart
df.sort_values(['Year', 'Period'], inplace=True)
plt.step(df.index, df['Value'], where='post')

plt.xlabel('Data Point')
plt.ylabel('Value')
plt.title('Step Chart')
plt.xticks(df.index, df['Year'].astype(str) + '-' + df['Period'])
plt.show()
```
