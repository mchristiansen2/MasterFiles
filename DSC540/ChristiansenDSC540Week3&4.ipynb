```python
# read in descriptions and header names
names = []
with open('/Users/madisonchristiansen/Desktop/adult_income_names.txt','r') as f:
    for line in f:
        f.readline()
        var=line.split(":")[0]
        names.append(var)
        
names
```

```python
# add name income for the last cloumn
names.append('Income') # add name

incomedf = pd.read_csv('/Users/madisonchristiansen/Desktop/adult.data', names=names) # add to column in new df
incomedf.head()
```

```python
# summary - you can see only five columns where shown
incomedf.describe()
```

```python
# variable factors - loop to print
varfactors = ['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 'occupation', 'relationship', 'race', 'sex', 'capital-gain', 'capital-loss', 'hours-per-week', 'native-country']

# loop to print
for f in varfactors:
    classes=incomedf[f].unique()
    numbclasses = incomedf[f].nunique()
    print("There are {} classes in the \"{}\" column: {}".format(numbclasses,f,classes))
    print("-"*100)
```

```python
# finding missing data
incomedf.isnull().sum
```

```python
# subsetting
incomedf_sub = incomedf[['age','education','occupation','race']]
incomedf_sub.head()
```

```python
# histogram of age
incomedf_sub['age'].hist(bins=20)
```

```python
# boxplot of age by race
incomedf_sub.boxplot(column='age',by='race', figsize=(10,5))
plt.xticks()
plt.xlabel("Race")
plt.show()
```

```python
def stripspace(s):
    return s.strip() # strip white space

# apply, new column, copy values over, drop new column
# age column
incomedf_sub['education_stripped']=incomedf['education'].apply(stripspace)
incomedf_sub['education']=incomedf_sub['education_stripped']
incomedf_sub.drop(labels=['education_stripped'],axis=1,inplace=True)

# education column
incomedf_sub['education_stripped']=incomedf['education'].apply(stripspace)
incomedf_sub['education']=incomedf_sub['education_stripped']
incomedf_sub.drop(labels=['education_stripped'],axis=1,inplace=True)

# occupation column
incomedf_sub['occupation_stripped']=incomedf['occupation'].apply(stripspace)
incomedf_sub['occupation']=incomedf_sub['occupation_stripped']
incomedf_sub.drop(labels=['occupation_stripped'],axis=1,inplace=True)

# race column
incomedf_sub['race_stripped']=incomedf['race'].apply(stripspace)
incomedf_sub['race']=incomedf_sub['race_stripped']
incomedf_sub.drop(labels=['race_stripped'],axis=1,inplace=True)
```

```python
# How many black people of age between 30 and 50 (inclusive) are represented in this dataset?

filtered=incomedf_sub[(incomedf_sub['race']=='Black') & (incomedf_sub['age']>=30) & (incomedf_sub['age']<=50)]
filtered.head()
# element at 0 index
answer=filtered.shape[0]

print("There are {} black people of age between 30 and 50.".format(answer))
```

```python
# use groupby to see mean age distribution
incomedf_sub.groupby(['race','education']).mean()
```

```python
# groupby occupation to show age statistics 
occupationstats=incomedf_sub.groupby('occupation').describe()['age']

occupationstats
```

```python
# finding outlier for occupation
plt.figure(figsize=(10,5)) # figure to show outlier
plt.barh(y=occupationstats.index,width=occupationstats['count'])
plt.yticks()
plt.show()
```

```python
# merging two dataset based on one variable, occupation

dataset1 = incomedf[['age','education','occupation']].sample(5,random_state=101) # first disjoint dataset
dataset2 = incomedf[['sex','race','occupation']].sample(5,random_state=101) # second disjoint dataset
datasets_merged = pd.merge(dataset1,dataset2,on='occupation',how='inner').drop_duplicates() # merged
datasets_merged
```
