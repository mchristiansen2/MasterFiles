```python
import pandas as pd

alsdf = pd.read_csv('/Users/madisonchristiansen/Desktop/als_data.csv')   
alsdf.head()
```

```python
list(alsdf.columns)
```

```python
## Columns to remove - ID and SubjectID do not contirbute to the disease, 
# gender and age do not contribute when patient has confirmed als 
remove = ['ID', 'Age_mean', 'SubjectID','Gender_mean']
alsdf2 = alsdf.drop(columns=remove)
list(alsdf2.columns)
```

```python
from sklearn.preprocessing import StandardScaler
## Aapply a standard scaler
scaler = StandardScaler()
scaler.fit(alsdf2)

scaled_data = scaler.transform(alsdf2)
alsdf_scaled = pd.DataFrame(scaled_data, columns=alsdf2.columns)

print(alsdf_scaled)
```

```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt


# empty lists to store the cluster numbers and silhouette scores
cluster = []
silhouette_scores = []

for k in range(2,15):
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(scaled_data)
    silhouette_avg = silhouette_score(scaled_data, labels) # calc silhouette score
    cluster.append(k) # append to list
    silhouette_scores.append(silhouette_avg) # append to list


plt.plot(cluster, silhouette_scores)
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score vs Number of Clusters')
plt.show()
```

```python
## choose an optimal number of clusters for K-means
optimal_num = 5
## Fit a K-means model to the data with the optimal number of clusters
kmeans = KMeans(n_clusters=optimal_num, random_state=42)
clusterlab = kmeans.fit_predict(alsdf_scaled)
```

```python
from sklearn.decomposition import PCA
## Fit a PCA transformation with two features to the scaled data
pca = PCA(n_components=2)
pca_transformed = pca.fit_transform(alsdf_scaled)
```

```python
## Make a scatterplot of the PCA transformed data coloring each point by its cluster value
plt.scatter(pca_transformed[:, 0], pca_transformed[:, 1], c=clusterlab)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA Transformed Data with Cluster Labels')
plt.show()
```

### The removed data was choosen to allow for the focus to be on the patients ALS condition and not the individuals. The standard scalar allowed to ensure all the features are on a simlar scale. The plot made of the cluster silhouette scores vs the number of clusters measure the speration of the clusters. The optimal number of clusters is chosen on the highest silhouette score, which is seen as 5. The PCA scatter plot shows a visual repersentation of the seperation and distribution of the clusters. The results of the clustering data will help indentify groups or patterns within the patient data. 

### Based on the PCA transformed data with cluster labels you can see that the cluters are mixed and overlapping meaning that the cluster algorithm could not accurately seperate the data into distinct groups.
