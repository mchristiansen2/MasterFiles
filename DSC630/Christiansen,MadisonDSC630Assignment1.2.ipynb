```python
import pandas as pd
import matplotlib.pyplot as plt


ricedf = pd.read_csv('/Users/madisonchristiansen/Desktop/RICE.csv')   
ricedf.tail()
```

```python
## general histograms for all features 
for x in ricedf.columns[1:14]:
    plt.hist(ricedf[x])
    plt.title("Plot of "+ x)
    plt.show()
```

```python
## bar graph - most common pests 
ricedf['PEST NAME'].value_counts().plot(kind='bar')
```

```python
## bar graph - year and pest value
x = ricedf['Observation Year']
y = ricedf['Pest Value']
# plot
plt.bar(x,y,width=0.7,color=['pink'])
plt.xlabel('Observation Year')
plt.ylabel('Pest Value')
plt.title("Pest Value by the Year")
plt.show()
```

```python
## boxplot - years associated with pest numbers/type
ricedf.boxplot(column='Observation Year',by='PEST NAME', figsize=(20,10))
plt.ylabel("Observation Year")
plt.xlabel("PEST NAME")
plt.xticks()
plt.show()
```

```python
## bivariate plot - Pest Value and Collection Type
ricedf.plot(x='Location', y='Pest Value', kind='scatter', figsize=(10,10))
plt.show()
```

### Summary
### This small analysis shows how the data is layed out and will give a background of the data. The histograms shows a good overview of all of the features within the dataset. When looking at the bar graph of the year and pest values you can see that the collection/observation from 1990 to 2010 was the hightest. This is a good indication that these years are a good sample set for furture analysis of the data. The bar graph showing the pest values by the type shows that the most common pests observed were the Yellowstemborer. This indicates there are a high number of observations for this pest species. All these visuals give a good understadning and show a overview of the dataset which allows to see were the analysis needs to be taken futher.
