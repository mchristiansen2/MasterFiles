```python
# Merging both datasets for ratings and title
dfmerged = pd.merge(dfmovies, dfratings, how='left', on='movieId')
dfmerged.head()
```

```python
# Limiting dataset to top movies - kernal dies due to large dataset
topmovies = dfmerged['title'].value_counts().head(1000).index
dfmerged = dfmerged[dfmerged['title'].isin(topmovies)]
```
