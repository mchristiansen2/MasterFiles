```python
# grouping the data
lenghts = fishdata[['Species', 'Length1', 'Length2', 'Length3']]
lenghts
```

```python
# setting up graph
x = lenghts.Species
ay = lenghts.Length1
by = lenghts.Length2
cy = lenghts.Length3
y = np.vstack([ay, by, cy])
# plot
fig, ax = plt.subplots()
ax.stackplot(x, y)
ax.set(xlim=(0,6),
       ylim=(0,150))
plt.xlabel('Species')
plt.ylabel('Lengths')
plt.title("Lengths of the Fish Species")
plt.show()
```

#### This graph shows the different lengths of each fish species. You can see that the lengths are all verying with the length 1 and 2 being the smallest. 

```python
# width data
width = fishdata[['Species', 'Width']]
width
```

```python
# make data
x = width.Species
y = width.Width
# plot
plt.plot(x,y)
plt.xlabel('Species')
plt.ylabel('Width')
plt.title("Width of the Fish Species")

plt.show()
```

#### This graph  is showing the width of the fish species. The highest widths recorded are the Pike species and the smallest widths recorded are the Smelt species.
