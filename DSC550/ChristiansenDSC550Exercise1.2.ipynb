```python
# Loading the dataset as Pandas data frame
import pandas as pd
import numpy as np

sales_df = pd.read_csv('/Users/madisonchristiansen/Desktop/Video_Games_Sales_as_at_22_Dec_2016.csv')
```

```python
# Top five games by critic score

sales_df.nlargest(n=5, columns='Critic_Score')
```

```python
# the number of video games in the data frame in each genre

sales_df['Genre'].value_counts()
```

```python
# calculating the total global sales for each publisher

mean_sales = sales_df.groupby('Publisher').agg('mean')
```

```python
# five publisher with highest total global sales

mean_sales.nlargest(n=5, columns='Global_Sales')
```

```python
# creating a new column to figure out the sales precentages

new_salesdf = sales_df.assign(NA_SalesPrecent = (sales_df.NA_Sales/sales_df.Global_Sales)*100)
```

```python
# the number of nan in each column

new_salesdf.isna().sum()
```

```python
# switching the non-numeric to NaN
new_salesdf['User_Score'] = pd.to_numeric(new_salesdf['User_Score'], errors='coerce')

# finding the median 
new_salesdf['User_Score'].median()
```

```python
# Filling NaN with the median

new_salesdf['User_Score'] = new_salesdf['User_Score'].fillna(7.5)
```
