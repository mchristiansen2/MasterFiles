```python
## dropping the state and phone number columns 
# churndf2 = churndf2.drop(columns=['phone number'])  - not needed
# churndf2 = churndf2.drop(columns=['state']) - not needed when the area code is provided 
churndf2.head()
```

```python
## STEP 1 Changing the churn column to an int in order for it to be evaluated in the model 
## YES - 1, NO - 0
churndf2['churn'] = churndf2['churn'].astype(int)
churndf2.head()
```

```python
## STEP 2 - Changing the yes and no columns to int as well
churndf2['international plan'] = churndf2['international plan'].replace(to_replace=["yes","no"], value=[1,0]) # (Y=1 and N=0)
churndf2['voice mail plan'] = churndf2['voice mail plan'].replace(to_replace=["yes","no"], value=[1,0]) # (Y=1 and N=0)
churndf2.head()
```

### Using a Random Forest classifer model for the churn data. This type of model can capture the relationships and interactions between the various features in the dataset. Along with this it provides a measure of feature importance to identify the key factors contributing to churn. The metrics such as accuracy, percision, recall, and the F1 score will give a good understanding as to if the model is good for the data. 

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score


X = churndf2.drop('churn', axis=1) ## STEP 3 - choosing the churn as the target column 
y = churndf2['churn']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2) ## STEP 4 - training the data and setting the test size 

rf_classifier = RandomForestClassifier(n_estimators=100)
```

```python
## STEP 5 - train the classifier
rf_classifier.fit(X_train, y_train)
```

```python
## STEP 6 - predict on the test set
y_pred = rf_classifier.predict(X_test)

## STEP 7 - accuracy of the classifier
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
```

```python
# STEP 8 - precision
precision = precision_score(y_test, y_pred)
print(f"Precision: {precision}")

# STEP 9 - recall
recall = recall_score(y_test, y_pred)
print(f"Recall: {recall}")

# STEP 10 - F1 score
f1 = f1_score(y_test, y_pred)
print(f"F1 score: {f1}")
```

### The accuracy shows 96% which means that the Random Forest classifier correctly predicted the churn label for 96% of the instancess in the test set. The precision of 8% shows that out of the predictive positive only 8% were actual true positives. This is an indication that the models ability to correctly identify postivie instances is low. The recall and F1 score both tell this as well that the models preformace in precision and recall is poor. 

### Using a logistic regression model is efficent when looking at larger data sets. This type of model can interperte coefficients that can help understadn the impact of the different features on the churn likelihood. This model can be a good base line model that is simple and allow for a greater understanding. 

```python
## Trying a logistic regression model to check outcome and see if better
from sklearn.linear_model import LogisticRegression

logreg_model = LogisticRegression() 

logreg_model.fit(X_train, y_train)  # fitting the data

y_pred2 = logreg_model.predict(X_test) # predict 

# metrics
accuracy = accuracy_score(y_test, y_pred2)
precision = precision_score(y_test, y_pred2)
recall = recall_score(y_test, y_pred2)
f1 = f1_score(y_test, y_pred2)
```

```python
print(f"Accuracy: {accuracy}") 
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1 score: {f1}")
```

### With the accuracy being at 88% this is a good indication of the models ability to correctly predict the churn for 88% of the instances in the test set. The precision being 77% implies that the model has the ability to avoid the false postives. The Recall of 82% meanms that the model identified the positive churn results. Although the F1 scroe is still low the Logistic Regression model preformed better. For this data set the Logistic Regression model is a better fit.
