```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
from tensorflow import keras
from keras.datasets import mnist
```

```python
## MNIST data set
(X_train, y_train), (X_test, y_test) = mnist.load_data() # train
```

```python
## display the first five images 
fig, axs = plt.subplots(1, 5)
for i in range(5):
    axs[i].imshow(X_train[i])
    axs[i].set_title(str(y_train[i]))
    axs[i].axis('off')
plt.show()
```

```python
print("Training: {}".format(X_train.shape))
print("Testing: {}".format(y_test.shape))
```

```python
## preprocess 
X_train = X_train.reshape(-1, 28, 28, 1).astype('float32') / 255.0
X_test = X_test.reshape(-1, 28, 28, 1).astype('float32') / 255.0
y_train = keras.utils.to_categorical(y_train)
y_test = keras.utils.to_categorical(y_test)
```

```python
## build and train a Keras CNN classifier on the MNIST training set.
CNN = keras.Sequential([
    keras.layers.Conv2D(64, kernel_size=(5, 5), activation='relu', input_shape=(28, 28, 1)), # convolutional layer
    keras.layers.MaxPooling2D(pool_size=(2, 2)), # 2x2 window
    keras.layers.Flatten(),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(10, activation='softmax')])
CNN.compile(optimizer='rmsprop', 
            loss='categorical_crossentropy', 
            metrics=['accuracy'])
CNN.fit(X_train, y_train, batch_size=128)
```

```python
CNN.summary() # summary of model 
```

```python
## accuracy
test_loss, test_accuracy = CNN.evaluate(X_test, y_test)
print("Accuracy:", test_accuracy)
```

```python
## confusion matrix
y_pred = CNN.predict(X_test)
y_pred = np.argmax(y_pred, axis=1)
y_true = np.argmax(y_test, axis=1)
confusion_mat = confusion_matrix(y_true, y_pred)
```

```python
## summarize the results
print("Accuracy:", test_accuracy)
print("Confusion Matrix:", confusion_mat)
```
